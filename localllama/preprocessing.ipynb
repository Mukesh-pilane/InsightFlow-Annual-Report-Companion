```python
from pprint import pprint
```

```python
import fitz  # import package PyMuPDF

# Open some document, for example a PDF (could also be EPUB, XPS, etc.)
doc = fitz.open("jio.pdf")
```

```python
arg = doc.get_page_xobjects(13)
```

```python
pprint(arg)
```

```python
page = doc[89]  # this is the first page
```

```python
tabs = page.find_tables()
print(f"{len(tabs.tables)} table(s) on {page}")
pprint(tabs[0].extract())
```

```python
print(page.get_text("json"))
```

```python
pprint(page.get_drawings())
```
